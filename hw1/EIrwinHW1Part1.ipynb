```
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```
# functions :)

def max(x, y):
  return -y**2 + -x**2

def min(x, y):
  return y**2 + x**2

def horse(x, y):
  return y**4 - x**4
```

```
x = np.linspace(-3, 3, 15)
y = np.linspace(-3, 3, 15)
X, Y = np.meshgrid(x,y)
Z = max(X, Y)

plt.contour(X, Y, Z, 50)
plt.figure()

ax = plt.axes(projection='3d')
ax.grid(True)
ax.contour3D(X,Y,Z,50)
plt.figure()
```

```
x = np.linspace(-3, 3, 15)
y = np.linspace(-3, 3, 15)
X, Y = np.meshgrid(x,y)
Z = min(X, Y)

plt.contour(X, Y, Z, 50)
plt.figure()

ax = plt.axes(projection='3d')
ax.grid(True)
ax.contour3D(X,Y,Z,50)
plt.figure()
```
